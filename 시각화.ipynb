```python
x=np.random.rand(3000)
y=np.random.rand(3000)
plt.scatter(x,y)
(x*x + y*y <1).mean()*4
# 확률 구하기 - 점을 랜덤으로 찍었을 때 값이 1 이하인것의 확률 : 원주율 구하는거임
```

```python
plt.figure(figsize=(3,3))
plt.scatter(x,y, s=1)
xs = np.linspace(0,1,200)
ys = (1 - xs * xs)** (0.5)
plt.plot(xs, ys, c='r')
```
